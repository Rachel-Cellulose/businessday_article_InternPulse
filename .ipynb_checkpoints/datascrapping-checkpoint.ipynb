```python
import requests
from bs4 import BeautifulSoup
import time

# Base URL for pagination
base_url = "https://businessday.ng/tag/bdlead/page/"

# Number of pages to scrape (816 pages in your case)
total_pages = 816

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Loop through all pages
for page in range(1, total_pages + 1):
    url = f"{base_url}{page}/?amp"
    
    # Send the request to fetch the page
    response = requests.get(url, headers=headers)
    
    if response.status_code == 200:
        print(f"Successfully fetched page {page}")
        
        # Parse the page content
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find articles on the page
        articles = soup.find_all("div", class_="jeg_postblock_content")
        
        if not articles:
            print(f"No articles found on page {page}")
        
        # Loop through each article and extract title and content
        for article in articles:
            title_tag = article.find("h2", class_="post-title")
            if title_tag:
                title = title_tag.get_text(strip=True)
                link = title_tag.find("a")["href"]
                print(f"Title: {title}")
                print(f"Link: {link}")
                
                # Now, extract the full content from the article page
                article_response = requests.get(link, headers=headers)
                article_soup = BeautifulSoup(article_response.content, "html.parser")
                
                # Find and extract the article content
                content_div = article_soup.find("div", class_="post-content")
                if content_div:
                    paragraphs = content_div.find_all("p")
                    full_text = "\n\n".join(p.get_text() for p in paragraphs)
                    print(f"Content: {full_text}\n\n")
                else:
                    print("‚ùå Could not find content for this article.")
            else:
                print(f"No title found for article on page {page}")
    else:
        print(f"‚ùå Failed to fetch page {page}. Status code: {response.status_code}")
    
    # Optional: Add a delay between requests to avoid getting blocked
    time.sleep(2)
```

```python
import requests
from bs4 import BeautifulSoup

# Set the base URL (you can change this to whatever page you want to scrape from)
base_url = "https://businessday.ng/tag/bdlead/?amp"

# Define headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Function to scrape a single page
def scrape_page(url):
    print(f"üîé Scraping {url}...")

    # Send the GET request to the page
    response = requests.get(url, headers=headers)

    # Check if the request was successful
    if response.status_code != 200:
        print(f"‚ùå Failed to retrieve {url}")
        return []

    # Parse the page content with BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Find all articles on the page
    articles = soup.find_all("article")
    
    # List to hold article data
    article_links = []

    # Loop through the articles and extract the links
    for article in articles:
        link = article.find("a", href=True)
        if link:
            article_links.append(link["href"])

    return article_links

# Function to handle pagination and scrape multiple pages
def scrape_all_pages():
    all_links = []
    page = 1

    while True:
        # Construct the URL for each page
        page_url = f"{base_url}/page/{page}/?amp"
        
        # Scrape the page and get article links
        article_links = scrape_page(page_url)

        # If no articles were found, stop scraping
        if not article_links:
            print("‚úÖ No more articles found.")
            break

        # Add the found links to the list
        all_links.extend(article_links)
        
        print(f"‚úÖ Found {len(article_links)} articles on page {page}.")
        
        # Move to the next page
        page += 1

    return all_links

# Run the scraper
all_article_links = scrape_all_pages()

# Print all scraped links
print(f"‚úÖ Total articles found: {len(all_article_links)}")
for link in all_article_links:
    print(link)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# Base URL (you may need to adjust this based on the pagination structure)
base_url = "https://businessday.ng/business-economy/page/{}/?amp"

# Headers for the HTTP request
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Create and open the CSV file
with open("scraped_articles.csv", mode="w", newline='', encoding="utf-8") as file:
    writer = csv.writer(file)
    # Write header row
    writer.writerow(["Title", "Link", "Content"])
    
    # Define number of pages you want to scrape
    total_pages = 10  # Adjust this based on how many pages you want to scrape
    
    for page_num in range(1, total_pages + 1):
        # Construct the URL for the current page
        url = base_url.format(page_num)
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, "html.parser")
        
        # Find article links on the page
        article_links = soup.find_all("a", class_="post-title")  # Adjust the class accordingly

        # Loop through each article link
        for article in article_links:
            link = article.get("href")
            title = article.get_text(strip=True)

            # Get the article's content
            response = requests.get(link, headers=headers)
            article_soup = BeautifulSoup(response.content, "html.parser")
            content_div = article_soup.find("div", class_="post-content")  # Adjust class based on site structure
            if content_div:
                paragraphs = content_div.find_all("p")
                full_text = "\n\n".join(p.get_text() for p in paragraphs)
            else:
                full_text = "‚ùå Could not find the article content."

            # Write article data to CSV
            writer.writerow([title, link, full_text])

        print(f"Scraped page {page_num}.")

print("Scraping complete! Data saved to 'scraped_articles.csv'.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
}

base_url = "https://businessday.ng/business-economy/page/{}"
all_articles = []

for page in range(1, 11):
    url = base_url.format(page)
    print(f"üìÑ Scraping page {page}: {url}")
    
    try:
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code != 200:
            print(f"‚ùå Failed to fetch page {page} ‚Äî Status Code: {response.status_code}")
            continue

        soup = BeautifulSoup(response.content, "html.parser")
        articles = soup.select("h3.entry-title a")

        for article in articles:
            title = article.get_text(strip=True)
            link = article["href"]

            # Fetch article content
            article_resp = requests.get(link, headers=headers, timeout=10)
            article_soup = BeautifulSoup(article_resp.content, "html.parser")
            content_paragraphs = article_soup.select("div.td-post-content p")
            content = "\n".join(p.get_text(strip=True) for p in content_paragraphs if p.get_text(strip=True))

            all_articles.append({
                "Title": title,
                "Content": content
            })

            time.sleep(1)  # Gentle on the server

    except Exception as e:
        print(f"‚ùå Error on page {page}: {e}")

# Save to CSV
with open("scraped_articles.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["Title", "Content"])
    writer.writeheader()
    writer.writerows(all_articles)

print("‚úÖ Scraping complete. Data saved to 'scraped_articles.csv'")

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import csv

# Set up Chrome in headless mode (runs in the background)
options = Options()
options.add_argument('--headless')
options.add_argument('--disable-gpu')
options.add_argument('--no-sandbox')
options.add_argument('--window-size=1920,1080')
options.add_argument('--user-agent=Mozilla/5.0')

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

all_articles = []
base_url = "https://businessday.ng/business-economy/page/{}"

for page in range(1, 50):  
    print(f"üìÑ Scraping page {page}...")
    driver.get(base_url.format(page))

    try:
        # Wait for the page content to load (adjust the element to what‚Äôs on the page)
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "h3.entry-title a")))
        
        soup = BeautifulSoup(driver.page_source, "html.parser")
        links = soup.select("h3.entry-title a")

        # Debugging: Check how many links are found on this page
        print(f"Found {len(links)} articles on page {page}...")

        for link in links:
            title = link.get_text(strip=True)
            href = link["href"]

            driver.get(href)
            time.sleep(2)

            article_soup = BeautifulSoup(driver.page_source, "html.parser")
            paragraphs = article_soup.select("div.td-post-content p")
            content = "\n".join(p.get_text(strip=True) for p in paragraphs)

            all_articles.append({
                "Title": title,
                "Content": content
            })

    except Exception as e:
        print(f"Error on page {page}: {e}")

# Debugging: Check the number of articles collected
print(f"Total articles scraped: {len(all_articles)}")

# Save to CSV after scraping
with open("scraped_articles.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["Title", "Content"])
    writer.writeheader()  # Write headers only once
    writer.writerows(all_articles)

print("‚úÖ Scraping complete. Articles saved to 'scraped_articles.csv'")

# Quit the driver
driver.quit()
```
