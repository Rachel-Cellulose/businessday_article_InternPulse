```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/business_day_Cleaned_Data_Tokenized_Sentiment.csv')

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Grouping by Month (or Year-Month)
df['YearMonth'] = df['Date'].dt.to_period('M').astype(str)  #  Convert to string directly

# Average Polarity per Month
monthly_sentiment = df.groupby('YearMonth')['Polarity'].mean().reset_index()

plt.figure(figsize=(14,6))
sns.lineplot(data=monthly_sentiment, x='YearMonth', y='Polarity', marker='o')
plt.xticks(rotation=45)
plt.title('Average Sentiment Over Time (Monthly)', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Average Polarity', fontsize=14)
plt.grid(True)
plt.tight_layout()

plt.savefig('data/sentiment_over_time.png')

plt.show()
```
