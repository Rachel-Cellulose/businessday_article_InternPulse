```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('data/business_day_Cleaned_Data_Tokenized_Sentiment.csv')

# Polarity Score Distribution
plt.figure(figsize=(10,6))
sns.histplot(df['Polarity'], bins=30, kde=True, color='purple')
plt.title('Distribution of Article Polarity Scores', fontsize=16)
plt.xlabel('Polarity Score', fontsize=14)
plt.ylabel('Number of Articles', fontsize=14)
plt.grid(True)
plt.tight_layout()

plt.savefig('data/polarity_distribution.png')

plt.show()
```
