```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('data/business_day_Cleaned_Data_Tokenized_Sentiment.csv')


df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Year'] = df['Date'].dt.year


yearly_sentiment = df.groupby('Year')['Polarity'].mean().reset_index()


plt.figure(figsize=(12,7))
sns.lineplot(data=yearly_sentiment, x='Year', y='Polarity', marker='o', markersize=8, linewidth=2.5, color='#7f1d75')  # dark magenta line


for i in range(yearly_sentiment.shape[0]):
    plt.text(
        x=yearly_sentiment['Year'][i],
        y=yearly_sentiment['Polarity'][i] + 0.01,  # Slightly above each dot
        s=round(yearly_sentiment['Polarity'][i], 2),
        ha='center',
        fontsize=10,
        color='black'
    )


plt.xticks(yearly_sentiment['Year'], rotation=45)
plt.title('Average Sentiment Score Per Year', fontsize=18)
plt.xlabel

```
