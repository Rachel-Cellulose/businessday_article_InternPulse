```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('data/business_day_Cleaned_Data_Tokenized_Sentiment.csv')

# Top 5 Most Positive Articles
top_positive = df.sort_values(by='Polarity', ascending=False).head(5)

# Top 5 Most Negative Articles
top_negative = df.sort_values(by='Polarity', ascending=True).head(5)

plt.figure(figsize=(12,8))  
sns.barplot(data=top_positive, y='Title', x='Polarity', palette='Greens')
plt.title('Top 5 Most Positive Articles', fontsize=18)
plt.xlabel('Polarity Score', fontsize=16)
plt.ylabel('Article Title', fontsize=16)
plt.tight_layout()
plt.savefig('data/top_positive_articles_clear.png', dpi=300)
plt.show()

plt.figure(figsize=(8,6))  
sns.barplot(data=top_negative, y='Title', x='Polarity', palette='Reds_r')
plt.title('Top 5 Most Negative Articles', fontsize=18)
plt.xlabel('Polarity Score', fontsize=16)
plt.ylabel('Article Title', fontsize=16)
plt.tight_layout()
plt.savefig('data/top_negative_articles_clear.png', dpi=300)  
plt.show()
```
